```python
from langchain.agents import tool
```

```python

#Here we are using the tool decorator to convert the normal function into a tool that can be used by the agent. 
#The tool decorator takes care of the input and output formatting, and also allows us to add a description to the tool which can be used by the agent to understand what the tool does.
@tool
def multiply(a: int, b: int) -> int:
    """Multiplies two numbers."""
    return a * b
```

```python
multiply.invoke({"a":5, "b":10})
```

```python
multiply.name
```

```python
multiply.description
```

```python
multiply.args
```

```python
#Another example of a tool that returns the length of a word.
def get_word_length(word:str)->int:
    return len(word)
```

```python
get_word_length("Tushar Nirkulwar")
```

```python
get_word_length.invoke 
# here we can see that the normal function does not have the invoke method which is used to call the tool, 
# but after using the tool decorator, we can see that the function has the invoke method which can be used to call the tool.
```

```python
@tool
def get_word_length(word:str)->int:
    """Returns the length of the given word."""
    return len(word)
```

```python
get_word_length.invoke({"word":"Tushar Nirkulwar"})
```
